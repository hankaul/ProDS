```python
def outliers_iqr(data):
    q1,q3 = np.percentile(data,[25,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    return np.where((data>upper_bound)|(data<lower_bound))
```

```python
from scipy import stats

#등분산 검정
stats.levene(y_gu.Distance, m_gu.Distance)
# 값(statistic) > 0.05 이면 등분산
```

> 영등포의 평균이 마포구의 평균보다 다소 낮게 나와, 통계적으로 발생한 차이인지 우연히 발생한 차이인지 확인

> 두 개 그룹의 데이터에서 평균계산 > 같은 모집단에 나온 다른 샘플의 평균이 통계적으로 의미 있게 비슷한 값인지 판단

> 가설검정과 T분포 이용

> 귀무가설 : 마포구와 영등포구의 공유자전거 이동거리의 평균은 같다

> 대립가설 : 마포구와 영등포구의 공유자전거 이동거리는 같지 않다.

> p-value : 귀무가설이 참이라는 가정 하에 현재 두 평균값사이에 나타날 확률

```python
#이표본 t-test : 두개의 샘플데이터를 가지고 각각 평균을 구한다
stats.ttest_ind(y_gu.Distance, m_gu.Distance, equal_var = True)
#Levene의 등분산 검정에서 등분산으로 판정했으므로 equl_var = True
# statistic=-4.002195758414915, pvalue=6.298774059911862e-05
stats.ttest_ind?
```

### one sample test  : 한쪽은 알려진 값으로 평균값이 제시
### paired t- test : 학습법으로 공부하기 이전과 이후로 나누어 시험을 보고 점수에 변화가 있는가를 보는 방식 (평균 before-after)
### 이표본 t-test : 두개의 샘플데이터를 가지고 각각 평균을 구한다


## 분산 분석
> 3개 이상의 데이터 그룹을 비교하기 위해서 사용

> 2개 이상의 데이터 그룹도 가능

> 그룹간 분산 분석을 하는 이유
 
> 두 개의 평균이 통계적으로 다르다면 두 평균의 분산도 겹치지 않는다.

### 1. 등분산 분석 : 여러 그룹의 분산이 같은가를 확인
> 귀무가설 : 모든 그룹의 분산은 같다

> 대립가설 : 모든 그룹의 분산은 같은 것이 아니다

> 분산 분석을 하기위해선 귀무가설이 기각되지 않아야 한다.

### 1-1 귀무가설이 기각된다면

> 비모수적 분석 방법 사용 : 평균,분산 표준편차 등 통계량에 근거한 추론 방법을 사용하지 않는 것

> 데이터에 보완할 점 검토 (데이터양, 데이터 수집기간)

### 2. 분산 분석 

> 귀무 가설 : 모든 그룹의 평균이 같다

> 대립 가설 : 어떤 그룹의 평균은 같지 않다

> 분산석은 어떤 그룹이 다른지는 알려주지 않는다.

### 3. 사후 분석
