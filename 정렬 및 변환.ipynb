문제02. 가장 많은 데이터가 있는 세공 수준과 색상조합을 순서대로 고르면?

```python
dia = pd.read_csv('data2/diamonds.csv')
```

```python
dia_melt = pd.crosstab(dia.color,dia.cut).reset_index().melt(id_vars='color')
```

```python
dia_melt.iloc[[dia_melt['value'].idxmax()],:] #답 G,Ideal
```

문제 03. 세공수준별 색상별 가격과 케럿의 평균을 구하고 1캐럿당 가격이 가장 높은 세공수준과 색상 조합을 순서대로 고르면?

```python
dia_group = dia.groupby(['color','cut'])[['price','carat']].mean()
```

```python
dia_group['ratio'] = dia_group['price']/dia_group['carat']
```

```python
dia_group.sort_values(by='ratio',ascending=False) #답 Premium ,G
```
