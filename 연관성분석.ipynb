```python
import pandas as pd
from mlxtend import frequent_patterns
```

```python
df = pd.read_csv('data2/association_rules_mart.csv')
```

```python
df
```

```python
df = df.iloc[:,1:].drop_duplicates()
df
```

```python
len(df)
```

```python
df['purchase']=True
df_pivot = pd.pivot_table(df,index='ID',columns='Item',values='purchase',aggfunc=max)
```

```python
df_pivot.fillna(False,inplace=True)
```

```python
df_pivot
```

```python
item_freq = frequent_patterns.apriori(df_pivot, min_support = 0.005, use_colnames=True)
```

```python
item_freq
```

```python
df_rules = frequent_patterns.association_rules(item_freq, metric='lift',
                                                min_threshold = 1.5)
```

```python
df_rules.sort_values('lift',ascending=False)
```
