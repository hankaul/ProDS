### statsmodels - ols()
- ols() 함수내에 종속변수와 독립변수를 선언
- ols()함수의 fit()메서드로 모델 적합
- 변수명에 온점 등 특수문자가 있는 경우 오류 발생
- 모델 객체의 predict()메서드로 예측 -> 다른모델과 다르게 다른 변수들이있어도 학습된 변수로만 예측
- F 통계량 귀무가설 : 선형관계가 없다 /대립가설 : 선형관계가 있다.
- t 통계량 귀무가설 : 계수가 0이다 / 대립가설 : 계수가 0이 아니다.

## 로지스틱 회귀분석 특징
- 이항 로지스틱 회귀 분석은 종속 변수가 0과 1이며 베르누이 분포를 따를 경우 사용
- 모델의 산출 값은 각 데이터가 1이 될 확률이며 이진 분류를 위해서 경계값(threshold) 필요
- 모델 평가를 위해 각종 분류 관련 지표 및 AUC 활용

### statsmodels = Logit()
- 로지스틱 회귀분석을 실시하는 statsmodels의 함수
- endog, exog 인자에 각각 종속변수와 독립변수를 할당
- 산출 모델 객체의 params 어트리뷰트에 모델의 계수 저장
- 산출 모델 객체의 predict() 메서드로 예측값을 생산하며 이는 종속변수가 1이 될 확률값

```python
df = pd.read_csv('data2/diabetes.csv')
```

```python
df
```

```python
X_train, X_test = train_test_split(df, test_size =0.3, random_state=345)
```

```python
cols = ['Glucose', 'BloodPressure', 'Pregnancies', 'BMI', 'Age']
depth_list = [3,4,5,6]
accs=[]

for depth in depth_list:
    model = DecisionTreeClassifier(max_depth = depth)
    model.fit(X=X_train.loc[:,cols], y=X_train['Outcome'])
    y_pred = model.predict(X_test.loc[:,cols])
    accs = accs + [metrics.accuracy_score(y_pred = y_pred, y_true=X_test['Outcome'])]
```
