문제2. 여름과 겨울의 시간대별 registered 평균을 비교할 떄 가장 차이가 많이 나는 시각은?(bike)

```python
df = pd.read_csv('data2/bike.csv')
df.info()
```

```python
df['datetime']= pd.to_datetime(df['datetime'])
```

```python
df['time'] = df.datetime.dt.time
```

```python
df['time']
```

```python
df_group = df.groupby(['season','time'])['registered'].mean().reset_index()
```

```python
df_group2 = df_group[df_group['season']==2]
df_group2
```

```python
df_group4 = df_group[df_group['season']==4]
df_group4
```

```python
merge_df = pd.merge(df_group2,df_group4, on='time')
```

```python
merge_df.head()
```

```python
merge_df['diff'] = abs(merge_df['registered_x'] - merge_df['registered_y'])
```

```python
merge_df.iloc[[merge_df['diff'].idxmax()],:] #답 19시
```
